```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
plt.figure(figsize=(5,3),dpi=100)
plt.title("our first graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.plot([1,2,3,6,8,9],[2,3,6,5,4,8])
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
#plt.plot(x,y, label='x' , color='black', linewidth=2, marker='o', markersize=6, markeredgecolor='white',linestyle='--' )
plt.plot(x,y,'g+--', label='x')
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks([0,1,2,3,7,9])
plt.yticks([0,2,6,8])

plt.legend()

plt.show()
```

```python
x2=np.arange(0,4,0.5)
print(x2)
```

```python
plt.plot(x2,x2**2, 'r', label='x^2')
plt.legend()
plt.show()
```

```python
plt.plot(x2[:6],x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:],x2[5:]**2, 'r--')
plt.legend()

plt.savefig('mygraph.png',dpi=200)
plt.show()
```

# Bar charts

```python
labels= ['A','B','C']
values=[1,4,2]

bars=plt.bar(labels,values)
#bars[0].set_hatch('/')
#bars[1].set_hatch('+')
#bars[2].set_hatch('*')

patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(6,3))

plt.show()
```
